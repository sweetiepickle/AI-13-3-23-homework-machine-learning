linear regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

url = 'linear regression.csv'
data= pd.read_csv(url)

import seaborn as sns
sns.heatmap(data.corr(), annot= True)   
#data.hist()
#plt.show()
```

```python
data.plot(kind='scatter', x='GPA', y='SAT', figsize= (9,9), color= 'black', xlim = (data['GPA'].min(),data['GPA'].max()))
```

```python
# hồi quy bậc 5

# trục x là GPA
# trục y là SAT

from sklearn.linear_model import LinearRegression
model= LinearRegression()
X= pd.DataFrame([data['GPA'],data['GPA']**2, data['GPA']**3, data['GPA']**4, data['GPA']**5, data['GPA']**6, data['GPA']**7, data['GPA']**8]).T   
#data['fixed acidity']= data['fixed acidity'].notna()    # Don't drop, just take the rows where fixed acidity is not NAN
model.fit(X, y= data['SAT'])
print('Intercept của mô hình: ', model.intercept_)
print('Coeffient của mô hình: ', model.coef_)
print('Độ chính xác của mô hình: ', model.score(X, data['SAT']))   
```

```python
range= np.arange(data['GPA'].min(), data['GPA'].max(), 0.1)
data.plot(kind='scatter', x='GPA', y='SAT', figsize= (9,9), color= 'black', xlim = (data['GPA'].min(),data['GPA'].max()))
predictor= pd.DataFrame([range, range**2, range**3, range**4, range**5, range**6, range**7, range**8]).T
y_pred= model.predict(X= predictor)
plt.plot(range, y_pred, color= 'blue')
```

Lung cancer

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

url = 'lung_cancer_examples.csv'
data= pd.read_csv(url)

import seaborn as sns
sns.heatmap(data.corr(), annot= True)   
```

```python
# hồi quy bậc 5
# trục x là result
# trục y là age
from sklearn.linear_model import LinearRegression
model= LinearRegression()
X=pd.DataFrame([data['Result'],data['Result']**2, data['Result']**3, data['Result']**4, data['Result']**5]).T   
model.fit(X, y=data['Age'])
print('Intercept của mô hình: ', model.intercept_)
print('Coeffient của mô hình: ', model.coef_)
print('Độ chính xác của mô hình: ', model.score(X, y= data['Age']))   
```

```python
data.plot(kind='scatter', x='Result', y='Age', figsize= (9,9), color= 'black', xlim = (data['Result'].min(),data['Result'].max()))
```

```python
range= np.arange(data['Result'].min(), data['Result'].max(), 0.08)
data.plot(kind='scatter', x='Result', y='Age', figsize= (9,9), color= 'black', xlim = (data['Result'].min(),data['Result'].max()))
predictor= pd.DataFrame([range, range**2, range**3, range**4, range**5]).T
y_pred= model.predict(X= predictor)
plt.plot(range, y_pred, color= 'blue')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

url = 'passenger.csv'
data= pd.read_csv(url)

import seaborn as sns
sns.heatmap(data.corr(), annot= True)   
```

```python
# hồi quy bậc 5
# trục x là miles
# trục y là passenger
from sklearn.linear_model import LinearRegression
model= LinearRegression()
X=pd.DataFrame([data['Miles(X)'],data['Miles(X)']**2, data['Miles(X)']**3, data['Miles(X)']**4, data['Miles(X)']**5, data['Miles(X)']**6, data['Miles(X)']**7]).T   
model.fit(X, y=data['Passenger'])
print('Intercept của mô hình: ', model.intercept_)
print('Coeffient của mô hình: ', model.coef_)
print('Độ chính xác của mô hình: ', model.score(X, y= data['Passenger']))   
```

```python
data.plot(kind='scatter', x='Miles(X)', y='Passenger', figsize= (9,9), color= 'black', xlim = (data['Miles(X)'].min(),data['Miles(X)'].max()))
```

```python
range= np.arange(data['Miles(X)'].min(), data['Miles(X)'].max(), 1)
data.plot(kind='scatter', x='Miles(X)', y='Passenger', figsize= (9,9), color= 'black', xlim = (data['Miles(X)'].min(),data['Miles(X)'].max()))
predictor= pd.DataFrame([range, range**2, range**3, range**4, range**5, range**6 ,range**7]).T
y_pred= model.predict(X= predictor)
plt.plot(range, y_pred, color= 'blue')
```
